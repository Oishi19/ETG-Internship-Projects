## Adding New Item into Inventory

```python
product_id = str(input("Enter the product id:"))
item = str(input("Enter name of the item:"))
price = int(input("Enter the price:"))
quantity = int(input("Enter the quantity:"))
category = str(input("Enter the item category"))
expiry = int(input("Enter the expiry in months:"))

records[product_id] = {'item': item, 'price': price, 'quantity': quantity, 'category': category, 'expiry in months': expiry}

js_file = json.dumps(records)

fd = open("records.json",'w')
fd.write(js_file)
fd.close()
```
